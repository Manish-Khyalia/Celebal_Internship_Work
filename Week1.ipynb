### Handling missing values

```python
# for numeric column

df['discount'] = df['discount'].fillna(0)

# for text columns

text_cols = [
    'what_customers_said',
    'seller_name',
    'videos',
    'seller_information',
    'variations'
]

df[text_cols] = df[text_cols].fillna('Unknown')
```

### Verifying missing values

```python
df.isnull().sum()
```

### Selecting  important columns

```python
df[['title', 'rating', 'initial_price', 'final_price']]
```

### Filtering rows

```python
df[df['rating'] > 4]
```

### Checking duplicate rows

```python
df.duplicated().sum()
```

### Removing duplicate rows

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

### Creating derived column

```python
# create quantity column
df['quantity'] = 2

# create total amount column
df['total_amount'] = df['initial_price'] * df['quantity']
```

### Displaying derived columns

```python
df[['initial_price', 'quantity', 'total_amount']].head()
```

### Saving cleaned dataset

```python
df.to_csv('cleaned_dataset.csv', index = False)
```
